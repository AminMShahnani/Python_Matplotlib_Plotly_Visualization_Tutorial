# Step Plot
### To analyze the Y-axis value with respect to Exactly X-Axis value

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(10, 5))
plt.style.use('Solarize_Light2')
x=np.arange(10)+0.5
y=np.random.uniform(1, 10, len(x))
print(x)
print(y)

# Plot
fig, ax=plt.subplots()
ax.step(x, y)
ax.set(xlim=(0, 10), xticks=np.arange(1, 10), ylim=(0, 10), yticks=np.arange(1, 10))
plt.show()
```

```python
fig=plt.figure(figsize=(10, 5))
x=np.linspace(0.1, 3 * np.pi, 50)
y=np.exp(np.sin(x))

plt.step(x, y)
plt.show()
```
