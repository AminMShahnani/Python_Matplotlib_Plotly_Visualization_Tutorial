# Pie Chart
### Pie Chart is circular graph or Chart which is divided into slices to represent numerical proportion. Each slice or part represent the proportionate part of the whole.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Animals=['Elephant', 'Tiger', 'Lion', 'Deer', 'Crocodiles']
Counts=[10, 32, 23, 40, 20]
fig=plt.figure(figsize=(10, 5))
plt.pie(Counts, labels=Animals, autopct='%1.2f%%')
plt.title('Zoo Pie Chart')
plt.show()
```

```python
Animals=['Elephant', 'Tiger', 'Lion', 'Deer', 'Crocodiles']
Counts=[10, 32, 23, 40, 20]
fig=plt.figure(figsize=(10, 5))
plt.pie(Counts, labels=Animals, autopct='%1.2f%%', startangle=90)
plt.title('Zoo Pie Chart')
plt.show()
```

```python
Animals=['Elephant', 'Tiger', 'Lion', 'Deer', 'Crocodiles']
Counts=[10, 32, 23, 40, 20]
Expld=[0, 0, 0.2, 0, 0]
fig=plt.figure(figsize=(10, 5))
plt.pie(Counts, labels=Animals, explode=Expld, autopct='%1.2f%%', startangle=90)
plt.title('Zoo Pie Chart')
plt.show()
```

```python
Animals=['Elephant', 'Tiger', 'Lion', 'Deer', 'Crocodiles']
Counts=[10, 32, 23, 40, 20]
Expld=[0, 0, 0.2, 0, 0]
fig=plt.figure(figsize=(10, 5))
plt.pie(Counts, labels=Animals, explode=Expld, shadow=True,autopct='%1.2f%%', startangle=90)
plt.title('Zoo Pie Chart')
plt.show()
```
