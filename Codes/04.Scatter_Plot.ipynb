# Scatter Plot
### Scatter Plot is a graph or diagram which uses dots or points to represent values for two different numerical variable. It is used to visualize the correlation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[10, 20, 30, 40, 50, 60, 70, 80]
p=[2, 3, 7, 5, 1, 4, 9, 2]
q=[6, 3, 2, 8,  5, 8, 3, 9]
fig=plt.figure(figsize=(12, 5))
S1=plt.subplot()
S1.scatter(x, p, color='g', s=300, edgecolor='r', marker='o')
S1.scatter(x, q, color='r', s=400, marker='4')
plt.title('Scatter  plot Example')
plt.grid(True)
plt.show()
```

```python
N=1000
X=np.random.rand(N)
Y=np.random.rand(N)
fig=plt.figure(figsize=(20, 10))
plt.scatter(X, Y, s=200, c='r', alpha=0.5)
plt.show()
```
