#  Stem Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(10, 5))
```

```python
#plt.style.use('_mpl-gallery')
plt.style.use('Solarize_Light2')
```

```python
np.random.seed(3)
x=np.arange(8)+0.5
y=np.random.uniform(2, 7, len(x))
```

```python
fig, ax=plt.subplots()
ax.stem(x,y)
ax.set(xlim=(0, 10), xticks=np.arange(1, 0), ylim=(0, 10), yticks=np.arange(1, 10))
plt.show()
```

```python
fig=plt.figure(figsize=(10, 5))
x=np.linspace(0.1, 3 * np.pi, 50)
y=np.exp(np.sin(x))
plt.stem(x,y)
plt.show()
```
