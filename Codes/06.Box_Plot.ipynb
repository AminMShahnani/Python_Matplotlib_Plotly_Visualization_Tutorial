# Box Plot

### 1.Box plot Summarize the dataset in efficient way without showing the individual points.
### 2.Box plot gives Minimum, Maximum, Median, 1st Quartile and Third Quartile

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Total_Sale=[20, 40, 30, 10, 25, 60, 120, 150, 50, 40, 30, 10]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.boxplot(Total_Sale)
plt.title('Summary of Sale')
plt.grid(True)
plt.show()
```

```python
Total_Sale=[20, 40, 30, 10, 25, 60, 120, 150, 50, 40, 30, 10]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.boxplot(Total_Sale, showmeans=True)
plt.title('Summary of Sale')
plt.grid(True)
plt.show()
```
