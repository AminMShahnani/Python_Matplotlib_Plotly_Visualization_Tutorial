# errorbar
### Plotly vs x as lines and/or markers with given errorbars.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(10, 5))
x=[2, 4, 6, 8, 10, 12]
y=[3, 6.7, 2.3, 4.5, 6.2, 2.3]
err=[0.3, 0.5, 0.6, 0.2, 0.1, 0.4]
plt.errorbar(x,  y, err, marker='s', mfc='red', mec='green', ms=5, mew=2)
plt.show()
```

```python
fig=plt.figure(figsize=(10, 5))
x=[2, 4, 6, 8, 10, 12]
y=[3, 6.7, 2.3, 4.5, 6.2, 2.3]
z=[0.3, 0.5, 0.6, 0.2, 0.1, 0.4]
plt.errorbar(x,  y, z, fmt='o', linewidth=2, capsize=6)
plt.show()
```
