```python
Roll=[1, 2, 3, 4, 5, 6]
Marks=[67, 78, 45, 34, 90, 89]
fig=plt.figure(figsize=(10, 5))
plt.xlabel('Roll Number')
plt.ylabel('Marks')
plt.title('Examination Data')
plt.plot(Roll, Marks, ls=':', marker='o', color='g')
plt.show()
```

```python
Roll_A=[1, 2, 3, 4, 5, 6]
Marks_A=[67, 78, 45, 34, 90, 89]
Roll_B=[1, 2, 3, 4, 5, 6]
Marks_B=[56, 23, 89, 65, 34, 90]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot(211) #2X1 1st Plot
S2=plt.subplot(212) #2X1 2nd Plot
S1.plot(Roll_A, Marks_A, ls=':', marker='o', color='r')
S2.plot(Roll_B, Marks_B, ls=':', marker='o', color='g')
plt.xlabel('Roll Number')
plt.ylabel('Marks')
S1.set(title='Section A')
S2.set(title='Section B')
plt.savefig('Fig_cmp')
plt.show()
```
