# Bar Plotting
### Bar Plot or Bar Chart is used to represent categorical data. The bars can be plotted vertically as well as horizontally. Example: Sales Data of Last 5 Years 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.bar(Month, Sales)
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.bar(Month, Sales)
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(Month, rotation='vertical')
plt.show()
```

```python
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.bar(Month, Sales, color='r')
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(Month, rotation='vertical', color='red')
plt.show()
```

```python
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.bar(Month, Sales, color='r', width=0.5)
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(Month, rotation='vertical', color='red')
plt.show()
```

```python
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Sales=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
fig=plt.figure(figsize=(10, 5))
S1=plt.subplot()
S1.barh(Month, Sales, color='r')
plt.title('Sales Data')
plt.xlabel('Sales')
plt.ylabel('Months')
plt.yticks(Month, color='red')
plt.grid(True)
plt.show()
```

```python
bw=0.20
Month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Week_1=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
Week_2=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
Week_3=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
Week_4=[1500, 3456, 7890, 7800, 6700, 3400, 2300, 9800, 6543, 1200, 1300, 4500]
w1=np.arange(len(Week_1))
w2=[x + bw for x in w1]
w3=[x + bw for x in w2]
w4=[x + bw for x in w3]
S1=plt.subplots(figsize=(30, 10))
plt.bar(w1, Week_1, color='r', width=bw, label='Week 1')
plt.bar(w2, Week_2, color='g', width=bw, label='Week 2')
plt.bar(w3, Week_3, color='b', width=bw, label='Week 3')
plt.bar(w4, Week_4, color='y', width=bw, label='Week 4')
plt.title('Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks([x + 0.25 for x in np.arange(len(Week_1))], Month)
plt.legend()
plt.show()
```
